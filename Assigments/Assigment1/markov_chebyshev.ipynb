```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 1000000
p = 0.5
k = 20
exp = np.random.binomial(k, p, n)
data = exp[exp*(1/20) >= 0.5]*(1/20)
```

```python
alphas = [0.5+0.05*x for x in range(11)]
_, ax = plt.subplots()
ax.hist(data, weights=np.zeros_like(data) + 1. / data.size)
y_markov = p/np.array(alphas)
y_chebyshev = (p*(1-p))/(np.array(alphas)**2)
ax.plot(alphas, y_markov, label='Markov')
ax.plot(alphas, y_chebyshev, 'b', label='Chebyshev')
plt.legend()
ax.set_ylabel('Empirical Frequencies')
ax.set_xlabel('Alpha')
ax.set_title('Empirical Frequencies w/ Markov/Chebyshev bounds.')
plt.show()
```
