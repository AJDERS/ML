## Import Modules

```python
import numpy as np
from matplotlib import pyplot as plt
import KNN
```

## Import Training Set

```python
picture_file = open('MNIST-cropped-txt/MNIST-Train-cropped.txt')
picture_content = picture_file.read().split()
picture_matrix = np.array(picture_content, dtype=float).reshape(10000, 784)

label_file = open('MNIST-cropped-txt/MNIST-Train-Labels-cropped.txt')
labels = np.array(label_file.read().split(), dtype=int)
```

## Import Test Set

```python
picture_file_test = open('MNIST-cropped-txt/MNIST-Test-cropped.txt')
picture_content_test = picture_file_test.read().split()
picture_matrix_test = np.array(picture_content_test, dtype=float).reshape(2000, 784)

label_file_test = open('MNIST-cropped-txt/MNIST-Test-Labels-cropped.txt')
labels_test = np.array(label_file_test.read().split(), dtype=int)
```

## Preprocessing

```python

p = KNN.Preprocess(picture_matrix, labels)
p_test = KNN.Preprocess(picture_matrix_test, labels_test)
```

```python
# Training/Validation set 0, 1
picture_matrix_0_1, labels_0_1 = p.pick_digits(0, 1)
data_train, data_val, labels_train, labels_val = p.make_train_val(picture_matrix_0_1, labels_0_1)
knn = KNN.Validation(data_train, data_val, labels_train, labels_val)
knn.validation_error_plot()
print(knn.best)

# Test set 0, 1 
picture_matrix_0_1_test, labels_0_1_test = p_test.pick_digits(0, 1)
knn = KNN.Validation(data_train, picture_matrix_0_1_test, labels_train, labels_0_1_test)
knn.validation_error_plot()
print(knn.best)
```

```python
# Training/Validation set 5, 6
picture_matrix_5_6, labels_5_6 = p.pick_digits(5, 6)
data_train, data_val, labels_train, labels_val = p.make_train_val(picture_matrix_5_6, labels_5_6)
knn = KNN.Validation(data_train, data_val, labels_train, labels_val)
knn.validation_error_plot()
print(knn.best)

# Test set 5, 6 
picture_matrix_5_6_test, labels_5_6_test = p_test.pick_digits(5, 6)
knn = KNN.Validation(data_train, picture_matrix_5_6_test, labels_train, labels_5_6_test)
knn.validation_error_plot()
print(knn.best)
```

```python
# Training/Validation set 0, 8
picture_matrix_0_8, labels_0_8 = p.pick_digits(0, 8)
data_train, data_val, labels_train, labels_val = p.make_train_val(picture_matrix_0_8, labels_0_8)
knn = KNN.Validation(data_train, data_val, labels_train, labels_val)
knn.validation_error_plot()
print(knn.best)

# Test set 0, 8 
picture_matrix_0_8_test, labels_0_8_test = p_test.pick_digits(0, 8)
knn = KNN.Validation(data_train, picture_matrix_0_8_test, labels_train, labels_0_8_test)
knn.validation_error_plot()
print(knn.best)
```
