```python
import numpy as np
import matplotlib.pyplot as plt
from linreg import LinReg
```

```python
f = open('DanWood.dt', 'r')
pairs =  f.read().split('\n')[:-1]
xs = [float(x.split()[0]) for x in pairs]
ys = [float(x.split()[1]) for x in pairs]
```

```python
lg = LinReg(xs, ys)
lg.plot()
lg.mse()/np.var(lg.ys)
```

```python
transform = lambda x: x**3
lg = LinReg(xs, ys)
lg.plot(transform)
lg.mse(transform)/np.var(lg.ys)
```
